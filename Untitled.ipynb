# Task 1: Understanding the Data

## 1.1 Dataset Description and Source

### Dataset Link: 
https://archive.ics.uci.edu/dataset/544/estimation+of+obesity+levels+based+on+eating+habits+and+physical+condition

### Dataset Overview:
The dataset contains 2,111 records in the collection, each with 17 attributes (16 features and 1 target variable). These records were taken from people in Mexico, Peru, and Colombia. The data includes information about both eating habits and health indicators. Notably, 77% of the data was created utilizing the Weka tool with the SMOTE filter to balance class distributions. The other 23% was collected directly from consumers through a web platform.

### Features Description:

***Physical Characteristics***:

- ***Gender***: Categorical (Male/Female)
- ***Age***: Continuous numerical (14-61 years)
- ***Height***: Continuous numerical (1.45-1.98 meters)
- ***Weight***: Continuous numerical (39-143 kg)

***Eating Habits***:

- ***FAVC*** (Frequent consumption of high caloric food): Binary (Yes/No)
- ***FCVC*** (Frequency of consumption of vegetables): Integer/Ordinal (Never/Sometimes/Always)
- ***NCP*** (Number of main meals): Continuous (1-2, 3, or >3)
- ***CAEC*** (Consumption of food between meals): Categorical (No/Sometimes/Frequently/Always)
- ***CH2O*** (Consumption of water daily): Ordinal (<1L, 1-2L, >2L)
- ***CALC*** (Consumption of alcohol): Categorical (I do not drink/Sometimes/Frequently/Always)

***Physical Condition & Lifestyle***: 

- ***family_history_with_overweight***: Binary (Yes/No)
- ***SCC*** (Calories consumption monitoring): Binary (Yes/No)
- ***FAF*** (Physical activity frequency): Ordinal (0-4 or 5 days)
- ***TUE*** (Time using technology devices): Ordinal (0-2h, 3-5h, >5h)
- ***SMOKE***: Binary (Yes/No)
- ***MTRANS*** (Transportation used): Categorical (Automobile/Motorbike/Bike/Public Transportation/Walking)

***Target Variable***:

- ***NObeyesdad*** (Obesity Level): Categorical with 7 classes - Insufficient Weight, Normal Weight, Overweight Level I, Overweight Level II, Obesity Type I, Obesity Type II, Obesity Type III

## 1.2 Classification Task Justification

### Classification Task:
The goal is to figure out how obese a person is (one of seven categories) based on their eating habits, physical traits, and way of life.  This is a multi-class classification problem. The goal variable (NObeyesdad) is categorical and has ordered classifications that go from underweight to severely obese.

### Why Decision Trees Are Reasonable:
- ***Interpretability***: Healthcare professionals, nutritionists, and patients can easily see and understand decision trees since they are clear, rule-based models. There are explicit "if-then" rules for each decision path that show how different lifestyle choices can lead to varying levels of obesity.

- ***Mixed Data Types***: The dataset has a mix of continuous features (like age, height, and weight), ordinal features (like how often you eat vegetables and how much you exercise), and categorical features (like gender and how you travel around).  Decision trees can work with different sorts of data without needing a lot of extra work.

- ***Non-linear Relationships***: The link between attributes (such BMI, which is based on height and weight) and levels of obesity is not linear.  Decision trees can model these intricate interactions without presuming linear correlations.

- ***Clinical Relevance***: Previous studies have shown that decision trees work quite well on this dataset, with precision rates of 97.4% and true positive rates of 97.8%.  The initial dataset paper effectively employed decision tree models.


- ***Feature Interactions***: Decision trees can find interactions between features (such eating a lot of high-calorie foods and not exercising much) that may have a bigger effect on obesity levels than any one feature on its own.

- ***Manageable Size***:The dataset is a good size for decision tree learning because it has 2,111 samples and 16 features. It is big enough to avoid serious overfitting but small enough to train quickly and read tree structures.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

csv_file = 'ObesityDataSet_raw_and_data_sinthetic.csv'
df = pd.read_csv(csv_file)
df.columns = [col.lower() for col in df.columns]

df['bmi'] = df['weight'] / (df['height'] ** 2)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Panel 1: Class distribution
sns.countplot(x='nobeyesdad', data=df, ax=axes[0], order=df['nobeyesdad'].value_counts().index)
axes[0].set_title('Class Distribution (Obesity Level)')
axes[0].set_xlabel('Obesity class')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=30)

# Panel 2: BMI by class (boxplot)
sns.boxplot(x='nobeyesdad', y='bmi', data=df, ax=axes[1], order=df['nobeyesdad'].value_counts().index)
axes[1].set_title('BMI Distribution by Obesity Level')
axes[1].set_xlabel('Obesity class')
axes[1].set_ylabel('BMI')
axes[1].tick_params(axis='x', rotation=30)

# Panel 3: Age by class (violinplot)
sns.violinplot(x='nobeyesdad', y='age', data=df, ax=axes[2], order=df['nobeyesdad'].value_counts().index, inner='quartile')
axes[2].set_title('Age Distribution by Obesity Level')
axes[2].set_xlabel('Obesity class')
axes[2].set_ylabel('Age')
axes[2].tick_params(axis='x', rotation=30)

num_features = ['age', 'height', 'weight', 'fcvc', 'ncp', 'ch2o', 'faf', 'tue']
# Compute correlation matrix
corr = df[num_features].corr()
# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Feature Correlation Heatmap: Obesity Dataset')

from pandas.plotting import parallel_coordinates

features = ['age', 'weight', 'height', 'faf', 'fcvc']
plot_df = df[features + ['nobeyesdad']].copy()

plt.figure(figsize=(14, 7))
parallel_coordinates(plot_df, class_column='nobeyesdad', color=sns.color_palette('tab10', n_colors=plot_df['nobeyesdad'].nunique()))
plt.title("Parallel Coordinates Plot: Obesity Numeric Features by Class")
plt.xlabel("Feature")
plt.ylabel("Normalized Value")
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1.0))

plt.tight_layout()
plt.show()
```
